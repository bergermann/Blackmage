```julia
using Pkg
Pkg.add(url="https://github.com/bergermann/Blackmage")
```

```julia
using Blackmage, JLD2
# use julias help mode to read information of each function
```

```julia
# connect to devices, insert respective IP and port number
#  MC -> motor control
# IDS -> Interferometric Displacement Sensor
# IDS web interface can be opened by putting the IP adress into any browser

mc_ips =    [
    ip"192.168.2.1",
    ip"192.168.2.2",
    ip"192.168.2.3",
]

ids_ips =   [
    ip"192.168.3.1",
    ip"192.168.3.2",
    ip"192.168.3.3",
]

md = MultiDevice(mc_ips,ids_ips)
md[1].settings.ess = (15e-6,15e-6,15e-6); md[1].settings.mrss = (10,10,10)
md[2].settings.ess = (10e-6,10e-6,10e-6); md[2].settings.mrss = (10,10,10)
```

```julia
# check port status
Blackmage.open_status(md)
```

```julia
# (re)open ports if they were closed
open(md)
```

```julia
# run this to close connection
close(md)
```

```julia
# check if interferometer measurements are enabled
getMeasurementEnabled(md,req)
```

```julia
# (re)start interferometer measurements (might take some time, see web interface)
startMeasurement(md,req)
```

```julia
# stop interferometer measurements
stopMeasurement(md,req)
```

```julia
# rezero interferometer axes
resetAxes(md,req)
```

WARNING: Emergency shutdown not yet properly tested/implemented. Discuss with Christoph what to do if anything goes wrong.

```julia
# run this once at start, after rezeroing interferometer axes
mcSetupFCM(md)
mcTargetFCM(md,[0.,0.,0.],:mm); mcWaitForTarget(md); sleep(1)
```

```julia
# target positions
mcSetupFCM(md)
mcTargetFCM(md,[1.,2.,3.],:mm); mcWaitForTarget(md); sleep(1)
```

```julia
# stop all motors and set them back to direct drive mode
mcStopAll(md)
```
